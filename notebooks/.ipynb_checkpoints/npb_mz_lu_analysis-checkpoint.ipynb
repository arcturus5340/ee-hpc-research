```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
file_path = "summary_annotated.csv"

df = pd.read_csv(
    file_path,
    delimiter=",",
    encoding="utf-8",
    header=0
)

# df.drop(columns=set(df.columns) - {x for x in df.columns if x[-4:] not in ["_min", "_max", "_std"]}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df_without_overflow = df[df["total_time"] < 19*60 + 50]
df_without_overflow.head()
```

```python
df_without_overflow.shape
```

```python
fig, axs = plt.subplots(2, 4, figsize=(25, 10))
first_step_runtime = None

for i, task in enumerate(df["taskspernode"].unique()):
    subset_df = df[(df["taskspernode"] == task)]
    axs[i//4, i%4].plot(subset_df['cpufreq'], subset_df['total_time'])
    axs[i//4, i%4].set_title(f'taskspernode={task}, threadspertask={96//task}')
    axs[i//4, i%4].set_xlabel('Frequency (Hz)')
    axs[i//4, i%4].set_ylabel('Execution Time (s)')
    axs[i//4, i%4].set_ylim([0, 1500])

plt.tight_layout()

plt.savefig("cpufreq_total_time.png", dpi=300)
plt.show()
```

```python
fig, axs = plt.subplots(2, 4, figsize=(25, 10))

for i, task in enumerate(df_without_overflow["taskspernode"].unique()):
    subset_df = df_without_overflow[(df_without_overflow["taskspernode"] == task)]

    total_energy_list = []
    for freq in subset_df["cpufreq"].unique():
        single_run_df = pd.read_csv(
            f"results/{task}_ranks_{96 // task}_threads_{freq}_hz.csv",
            delimiter=",",
            encoding="utf-8",
            header=0
        )

        total_energy_list.append(single_run_df["total_energy"].max())
        
    axs[i//4, i%4].plot(subset_df['cpufreq'], total_energy_list)
    axs[i//4, i%4].set_title(f'taskspernode={task}, threadspertask={96//task}')
    axs[i//4, i%4].set_xlabel('Frequency (Hz)')
    axs[i//4, i%4].set_ylabel('Energy Consumed (J)')
    axs[i//4, i%4].set_ylim([0, 250000])
    axs[i//4, i%4].set_xlim([0.9e6, 3.99e6])

plt.tight_layout()

plt.savefig("cpufreq_total_energy.png", dpi=300)
plt.show()
```
